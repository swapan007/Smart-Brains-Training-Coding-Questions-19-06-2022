```python
from math import gcd 
# Function to find maximum GCD
def max_gcd(l1):
    if (len(l1) == 1 or len(l1) == 0):
        return 0       
    l1.sort() 
    # To get the first pair
    a = l1[0]
    ans = -10**9
    for i in range(1, len(l1)):
        b = l1[i] 
        # If both the pairs belongs to the same parent
        if (b[0] == a[0]):
            ans = max(ans, gcd(a[1], b[1]))
        # for next iteration
        a = b
    return ans
#  main function starts from here
if __name__ == '__main__':
    l1=[]
    l1.append([4,5])
    l1.append([4,2])
    l1.append([2,3])
    l1.append([2,1])
    l1.append([3,6])
    l1.append([3,12])
 
    print(max_gcd(l1))
```

```python
def findElement(arr, n) :
	prefixSum = [0] * n
	prefixSum[0] = arr[0]
	for i in range(1, n) :
		prefixSum[i] = prefixSum[i - 1] + arr[i]
	suffixSum = [0] * n
	suffixSum[n - 1] = arr[n - 1]
	for i in range(n - 2, -1, -1) :
		suffixSum[i] = suffixSum[i + 1] + arr[i]
	for i in range(1, n - 1, 1) :
		if prefixSum[i] == suffixSum[i] :
			return i		
	return -1
# main code
if __name__ == "__main__" :
	
	arr = [1,2,3,4,6,1,2,3,4]
	n = len(arr)
	print(findElement(arr, n))
```

```python
def burst_balloon(arr):
    out = 0
    len_arr = len(arr)
    for i in range(len_arr):
        v = arr[i]
        if v == -1:
            continue;
        v -= 1
        out = out +1
        arr[i] = -1
        for j in range(i+1, len_arr):
            if arr[j] == v:
                arr[j] = -1
                v -= 1
    return out
 
A = [6, 5, 1, 4, 5]
out  = burst_balloon(A)
print(out)
'''
INPUT:
[6,5,1,4,5]

OUTPUT:
3
'''
```
